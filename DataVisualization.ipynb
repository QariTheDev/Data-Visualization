## Data Visualization


```python
#importing libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#import data
kashti = sns.load_dataset('titanic')
kashti.head(7)
```

```python
plt.plot(kashti['age'], kashti['fare'])
plt.show()
```

```python
plt.plot(kashti['age'], kashti['fare'], 'o')
plt.show()
```

```python
plt.plot(kashti['age'], kashti['fare'], 'o', color='red', alpha=0.5)
plt.show()
```

```python
#customize the plot
plt.plot(kashti['age'], kashti['fare'],'o', color = 'purple')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()
```

```python
#histogram
plt.hist(kashti['survived'], bins = 10)
plt.show()
```

```python
#value count
kashti.value_counts().sum()
```

```python
kashti['age'].value_counts().plot(kind='bar')
plt.show()
```
